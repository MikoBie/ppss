## Task 2. (4 points)

Write a function that takes a string and reverses every word in it, but keeps the words in the same order. It should return a string. Write a proper `docstring` for the function. It should work like the following. **Hint**: You might want to convert the string to a list.

```python
backward_string_by_word('') == ''
backward_string_by_word('rabarbar') == 'rabrabar'
backward_string_by_word('I <3 my kayak') == 'I 3< ym kayak'
backward_string_by_word('hello world') == 'olleh dlrow'
```

````python
def backward_string_by_word(str):
    """
    Takes a string and reverses every word in it, but keeps the words in the
    same order.

    Args:
        str (str): a string

    Returns:
        str : string in which every word is in reversed order compared to the
        initial string. The order of the words is the same as in the initial
        string
    """
    return " ".join([x[::-1] for x in str.split(" ")])


print(backward_string_by_word("") == "")
print(backward_string_by_word("rabarbar") == "rabrabar")
print(backward_string_by_word("I <3 my kayak") == "I 3< ym kayak")
print(backward_string_by_word("hello world") == "olleh dlrow")
````

## Task 3. (4 points)

Write a function that takes a list of numbers and returns a list without repetitions. Write a propoer `docstring` for the function.

```python
repetition_remover([1,2,3,4,5,5,1]) == [1,2,3,4,5]
```

````python
def repetition_remover(L1):
    """
    Takes a list of numeric values and returns a list without
    repetitions.

    Args:
        L1 (list): a list of numeric values

    Returns:
        list : a list of unique numeric values
    """
    result = []
    for item in L1:
        if item not in result:
            result.append(item)
    return result


repetition_remover([1, 2, 3, 4, 5, 5, 1]) == [1, 2, 3, 4, 5]
````
