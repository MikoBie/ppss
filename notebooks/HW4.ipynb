````python
def find_last_wrong(s, sub):
    """
    Returns the index of the last occurance of sub in s.
    Returns None if sub does not occur in s.

    Args:
        s (str): non-empty string
        sub (str): non-empty string

    Returns:
        (int/None): index of the last occurance of sub in s or.
        If sub does not occur in s returns None.
    """
    s_reversed = s[::-1]
    sub_reversed = sub[::-1]
    return len(s) - 1 - s_reversed.find(sub_reversed)
````

````python
def test_find_last(s_vals, sub_vals):
    """
    Tests the find_last function from HW3. It prints the value of s, sub, and the result of the test.

    Args:
            s_vals (tuple): A tuple of non-empty strings. 
            sub_vals (tuple): A tuple of non-empty strings.
    """
    for s in s_vals:
        for sub in sub_vals:
            result = find_last_wrong(s, sub)
            if result == s.rfind(sub):
                val = 'Ok'
            elif result == None and s.rfind(sub) == -1:
                val = 'Ok'
            else:
                val = 'Bad'
            print(f's = {s}, sub = {sub}: {val}')
            
s_vals = ('aba', 'ababab', 'a', 'aa')
sub_vals = ('a', 'ab', 'ababab', 'b', 'bb')

test_find_last(s_vals = s_vals, sub_vals = sub_vals) 
````
